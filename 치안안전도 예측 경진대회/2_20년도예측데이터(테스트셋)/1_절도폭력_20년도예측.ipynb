```python
# 20년도 통합본 불러오기
df_통합 = pd.read_csv("../20년도데이터/20년도_통합본.csv", index_col=0)
df_통합.head()
```

```python
#절도폭력 안전도 변수 불러오기
df = pd.read_csv("../완성본_안전도_회귀분석/1_절도폭력안전도.csv")
df = df[["관할서","연도","crm_clue_고소","crm_wthr_바람","crm_clue_현행범","crm_clue_피해자신고","vic_sx_2","vic_age_60세초과","cctv_개수",
         "배치인원_수","비상벨_개수",
        "일인가구수","기초수급_19세이하","외국인수", "score_절폭"]]
df.head()
```

```python
#20년도 데이터 구축위해 19년도만 가져오기
df_20 = df[df["연도"]==2019].reset_index(drop = True)
df_20
```

```python
#경찰청통계사이트에서 구한 19년도 대비 20년도 증감율을 곱하여 20년도 데이터를 예측한다
df_20["crm_clue_고소"] = np.round(df_20["crm_clue_고소"] *0.99777 ,0)
df_20["crm_wthr_바람"] = np.round(df_20["crm_wthr_바람"] *1.3,0)
df_20["crm_clue_현행범"] = np.round(df_20["crm_clue_현행범"] *0.73750 ,0)
df_20["crm_clue_피해자신고"] = np.round(df_20["crm_clue_피해자신고"] *0.95770 ,0)
df_20["vic_sx_2"] = np.round(df_20["vic_sx_2"] *0.92576   ,0)
df_20["vic_age_60세초과"] = np.round(df_20["vic_age_60세초과"] *1.00353   ,0)
```

```python
df_20
```

```python
# 20년도 데이터 있는 변수만 바꿔주기
df_20["비상벨_개수"]=df_통합["비상벨_개수"]
df_20["일인가구수"]=df_통합["일인가구수"]
df_20["기초수급_19세이하"]=df_통합["기초수급_19세이하"]
df_20["외국인수"]=df_통합["외국인수"]
```
