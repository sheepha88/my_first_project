```python
# 20년도 통합본 불러오기
df_통합 = pd.read_csv("../20년도데이터/20년도_통합본.csv", index_col=0)
df_통합.head(1)
```

```python
#강도살인 안전도 변수 불러오기
df = pd.read_csv("../완성본_안전도_회귀분석/2_강도살인안전도.csv")
df = df[["관할서","연도","crm_clue_변사체","crm_clue_자수","crm_clue_진정","crm_clue_현행범","crm_tm_저녁"
    ,"vic_sx_1","강도살인_검거수","화재_사망","화재_부상","화재_부동산피해_천원","공원_개수","자살_사망률_10만명당"
         ,"자살_연령표준화사망률_10만명당","기초수급_35_59세","기초수급_60세이상","외국인수", "score_강살"]]
df.head(1)
```

```python
#20년도 데이터 구축위해 19년도만 가져오기
df_20 = df[df["연도"]==2019].reset_index(drop = True)
df_20
```

```python
#경찰청통계사이트에서 구한 19년도 대비 20년도 증감율을 곱하여 20년도 데이터를 예측한다
df_20["crm_clue_변사체"] = np.round(df_20["crm_clue_변사체"] *0.92537  ,0)
df_20["crm_clue_자수"] = np.round(df_20["crm_clue_자수"] *0.95522   ,0)
df_20["crm_clue_진정"] = np.round(df_20["crm_clue_진정"] *1.20522   ,0)
df_20["crm_clue_현행범"] = np.round(df_20["crm_clue_현행범"] *0.71559    ,0)
df_20["crm_tm_저녁"] = np.round(df_20["crm_tm_저녁"] *0.91902    ,0)
df_20["vic_sx_1"] = np.round(df_20["vic_sx_1"] *0.91829     ,0)

```

```python
#강도살인검거수는 지역별로 분리가 가능하기때문에 지역별 증감율을 곱한다
for i in range(len(df_20)):
    if "경기" in df_20.loc[i,"관할서"]:
        df_20.loc[i,"강도살인_검거수"]=np.round(df_20.loc[i,"강도살인_검거수"]*0.96784 ,0)
        
    elif "경남" in df_20.loc[i,"관할서"]:
        df_20.loc[i,"강도살인_검거수"]=np.round(df_20.loc[i,"강도살인_검거수"]*0.99504  ,0)
        
    elif "서울" in df_20.loc[i,"관할서"]:
        df_20.loc[i,"강도살인_검거수"]=np.round(df_20.loc[i,"강도살인_검거수"]*0.91034   ,0)
        
    elif "충남" in df_20.loc[i,"관할서"]:
        df_20.loc[i,"강도살인_검거수"]=np.round(df_20.loc[i,"강도살인_검거수"]*1.36923    ,0)
```

```python
# 20년도 데이터 있는 변수만 바꿔주기
df_20["화재_사망"]=df_통합["화재_사망"]
df_20["화재_부상"]=df_통합["화재_부상"]
df_20["화재_부동산피해_천원"]=df_통합["화재_부동산피해_천원"]
df_20["기초수급_35_59세"]=df_통합["기초수급_35_59세"]
df_20["기초수급_60세이상"]=df_통합["기초수급_60세이상"]
df_20["외국인수"]=df_통합["외국인수"]
```
