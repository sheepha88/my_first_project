```python
# 20년도 통합본 불러오기
df_통합 = pd.read_csv("../20년도데이터/20년도_통합본.csv", index_col=0)
df_통합.head(1)
```

```python
#교통사고 안전도 변수 불러오기
df = pd.read_csv("../완성본_안전도_회귀분석/3_교통사고안전도.csv")
df = df[["관할서","연도","crm_wthr_눈","crm_clue_피해자신고","vic_age_2_30대","일인가구수"
    ,"총_인구수","기초수급_19세이하","기초수급_60세이상"
         ,"외국인수", "score_교통"]]
df.head(1)
```

```python
#20년도 데이터 구축위해 19년도만 가져오기
df_20 = df[df["연도"]==2019].reset_index(drop = True)
df_20.head(1)
```

```python
df_20["crm_wthr_눈"] = np.round(df_20["crm_wthr_눈"] *0.87  ,0)
df_20["crm_clue_피해자신고"] = np.round(df_20["crm_clue_피해자신고"] *0.94335  ,0)
df_20["vic_age_2_30대"] = np.round(df_20["vic_age_2_30대"] *1.00755   ,0)

```

```python
df_20
```

```python
#경찰청통계사이트에서 구한 19년도 대비 20년도 증감율을 곱하여 20년도 데이터를 예측한다
df_20["일인가구수"]=df_통합["일인가구수"]
df_20["총_인구수"]=df_통합["총_인구수"]
df_20["기초수급_19세이하"]=df_통합["기초수급_19세이하"]
df_20["기초수급_60세이상"]=df_통합["기초수급_60세이상"]
df_20["외국인수"]=df_통합["외국인수"]
```

```python
df_20["연도"]=2020
```
