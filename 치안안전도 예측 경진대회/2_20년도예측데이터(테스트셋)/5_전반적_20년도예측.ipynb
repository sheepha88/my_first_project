```python
# 20년도 통합본 불러오기
df_통합 = pd.read_csv("../20년도데이터/20년도_통합본.csv", index_col=0)
df_통합.head(1)
```

```python
#법질서 안전도 변수 불러오기
df = pd.read_csv("../완성본_안전도_회귀분석/5_전반적안전도.csv")
df = df[["관할서","연도","crm_wthr_흐림","crm_clue_타인신고","crm_clue_피해자신고","crm_clue_현행범","crm_tm_저녁","화재_사망","배치인원_수",
        "비상벨_개수","기초수급_19세이하","외국인수", "score_전반"]]
df.head(1)
```

```python
#20년도 데이터 구축위해 19년도만 가져오기
df_20 = df[df["연도"]==2019].reset_index(drop = True)
df_20.head(1)
```

```python
#경찰청통계사이트에서 구한 19년도 대비 20년도 증감율을 곱하여 20년도 데이터를 예측한다
df_20["crm_clue_타인신고"] = np.round(df_20["crm_clue_타인신고"] *1.02544   ,0)
df_20["crm_clue_피해자신고"] = np.round(df_20["crm_clue_피해자신고"] *0.97613    ,0)
df_20["crm_clue_현행범"] = np.round(df_20["crm_clue_현행범"] *0.79036    ,0)
df_20["crm_tm_저녁"] = np.round(df_20["crm_tm_저녁"] *0.98509     ,0)


```

```python
df_20
```

```python
# 20년도 데이터 있는 변수만 바꿔주기
df_20["화재_사망"]=df_통합["화재_사망"]
df_20["비상벨_개수"]=df_통합["비상벨_개수"]
df_20["기초수급_19세이하"]=df_통합["기초수급_19세이하"]
df_20["외국인수"]=df_통합["외국인수"]
```
