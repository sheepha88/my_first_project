```python
import pandas as pd
import numpy as np
```

```python
q1 = pd.read_csv('112신고 Q1.csv', index_col = 0)
q2 = pd.read_csv('112신고 Q2.csv', index_col = 0)
q3 = pd.read_csv('112신고 Q3.csv', index_col = 0)
q4 = pd.read_csv('112신고 Q4.csv', index_col = 0)
q5 = pd.read_csv('112신고 Q5.csv', index_col = 0)
```

```python
q1[q1.date==2018].head(10)
```

```python
q1_2018 = q1[q1.date==2018]
q1_2019 = q1[q1.date==2019]
q1_2018['q1 신고'] = q1_2018['q1 신고']*2


q2_2018 = q2[q1.date==2018]
q2_2019 = q2[q1.date==2019]
q2_2018['q2 신고'] = q2_2018['q2 신고']*2


q3_2018 = q3[q3.date==2018]
q3_2019 = q3[q3.date==2019]
q3_2018['q3 신고'] = q3_2018['q3 신고']*2


q4_2018 = q4[q4.date==2018]
q4_2019 = q4[q4.date==2019]
q4_2018['q4 신고'] = q4_2018['q4 신고']*2


q5_2018 = q5[q5.date==2018]
q5_2019 = q5[q5.date==2019]
q5_2018['q5 신고'] = q5_2018['q5 신고']*2
```

```python
q1_2018.head(10)
```

```python
q1_2017_value = np.round((q1_2018['q1 신고'] + q1_2019['q1 신고'])/2)
q1_2017 = q1_2018.copy()
q1_2017['q1 신고'] = q1_2017_value.astype(int)
q1_2017['date'] = 2017

q2_2017_value = np.round((q2_2018['q2 신고'] + q2_2019['q2 신고'])/2)
q2_2017 = q2_2018.copy()
q2_2017['q2 신고'] = q2_2017_value.astype(int)
q2_2017['date'] = 2017

q3_2017_value = np.round((q3_2018['q3 신고'] + q3_2019['q3 신고'])/2)
q3_2017 = q3_2018.copy()
q3_2017['q3 신고'] = q3_2017_value.astype(int)
q3_2017['date'] = 2017

q4_2017_value = np.round((q4_2018['q4 신고'] + q4_2019['q4 신고'])/2)
q4_2017 = q4_2018.copy()
q4_2017['q4 신고'] = q4_2017_value.astype(int)
q4_2017['date'] = 2017

q5_2017_value = np.round((q5_2018['q5 신고'] + q5_2019['q5 신고'])/2)
q5_2017 = q5_2018.copy()
q5_2017['q5 신고'] = q5_2017_value.astype(int)
q5_2017['date'] = 2017
```

```python
q1_2019.head(10)
```

```python
q1 = pd.concat([q1_2017, q1_2018, q1_2019])
q2 = pd.concat([q2_2017, q2_2018, q2_2019])
q3 = pd.concat([q3_2017, q3_2018, q3_2019])
q4 = pd.concat([q4_2017, q4_2018, q4_2019])
q5 = pd.concat([q5_2017, q5_2018, q5_2019])
```

```python
q1.to_csv('112신고 Q1 17년포함_csv')
q2.to_csv('112신고 Q2 17년포함_csv')
q3.to_csv('112신고 Q3 17년포함_csv')
q4.to_csv('112신고 Q4 17년포함_csv')
q5.to_csv('112신고 Q5 17년포함_csv')
```
