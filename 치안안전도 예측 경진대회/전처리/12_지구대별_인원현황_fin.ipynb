```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data/12.경찰서별_지구대별_인원현황.csv")
df.isnull().sum()
```

```python
df.head()
```

## 1. 전처리

```python
# 이상한 이름의 경찰서 확인-> 모두 서울구로 경찰서임
df[df["pol_office"].isin(['서울일지경찰서','서울척지경찰서', '서울봉지경찰서', 
                          '서울류지경찰서', '서울왕파경찰서', '서울로3경찰서', '서울리봉경찰서'])]
```

```python
# 다 구로경찰서로 바꾼다
df.loc[179:185,"pol_office"]="서울구로경찰서"
```

```python
# 관할서명  수정
df["pol_office"] = df["pol_office"].str.replace("마산", "경남마산")                                     
df["pol_office"] = df["pol_office"].str.replace("창원", "경남창원")
df["pol_office"] = df["pol_office"].str.replace("수원", "경기수원")
df["pol_office"] = df["pol_office"].str.replace("진주", "경남진주")
df["pol_office"] = df["pol_office"].str.replace("진해", "경남진해")
df["pol_office"] = df["pol_office"].str.replace("세종", "충남세종")                                           
```

```python
# 관할서별 인원 파악 후 데이터프레임 생성
df_pol_pop = df.groupby("pol_office", sort = True).sum("num").reset_index()
df_pol_pop
```

```python
#컬럼명 바꾸기
df_pol_pop.rename(columns = {"pol_office":"관할서","num":"pol_pop"}, inplace=True)
```

```python
#파일저장
df_pol_pop.to_csv("전처리완성본/12_관할서별_배치인원현황")
```
