## 2. 관할서 지정

```python
df_bl
```

```python
from tqdm import tqdm
```

```python
# 공원 위도경도가 관할경계에 들어가는지 파악하고, 들어간다면 관할서 지정해주는 코드
for i in tqdm(range(41)):
    for z in range(18334):
        if df_bl["geometry"][z].within(df_js["geometry"][i]):
            df_bl.loc[z,"관할서"]=df_js["NAME"][i]
            
```

## 3.전처리

```python
import pandas as pd
```

```python
df = pd.read_csv("../데이터별_관할서리스트/비상벨현황_관할서리스트.csv", index_col = 0)
df
```

```python
# 관할서명  수정
df["관할서"] = df["관할서"].str.replace("마산", "경남마산")                                     
df["관할서"] = df["관할서"].str.replace("창원", "경남창원")
df["관할서"] = df["관할서"].str.replace("수원", "경기수원")
df["관할서"] = df["관할서"].str.replace("진주", "경남진주")
df["관할서"] = df["관할서"].str.replace("진해", "경남진해")
df["관할서"] = df["관할서"].str.replace("세종", "충남세종")                                           
```

```python
df.isnull().sum()
```

```python
df[df["bell_set_yr"]==2020]
```

```python
df["bell_set_yr"].unique()
```

# 주의할점

+ 송파, 서초, 도봉 경찰서는 데이터가 누락되어있으므로 외부데이터 조사를 통해 값을 기입해야한다
+ 외부데이터가 없을 시, 임의로 기입한다

    + 서초경찰서는 17년도에 0개였을것으로 보인다 ->18년도에 1개 이기 때문에
    + 송파경찰서는 17년도에 815개, 18년도에 834, 19년도에 864, 20년도에 879 ->강남구 추이랑 비슷하게함
    + 도봉경찰서는 모든년도에 622 -> 강북, 노원구는 추이에 변화가 없다

## 17년도 비상벨 개수

```python
#17년도 비상벨 개수 : 총누적(20년도현황)- 20,19,18년도에 설치한 비상벨개수
# 송파, 서초, 도봉 경찰서가 없음 추가해야됨
#
df_17 = df[-df["bell_set_yr"].isin([2020,2019,2018])]
df_17
```

```python
#관할서별 비상벨 개수 데이터프레임화
df_bell_17 = pd.DataFrame(df_17.groupby("관할서", sort = True).count()["bell_us"]).reset_index()
df_bell_17.rename(columns={"bell_us":"bell_cnt"}, inplace=True)
df_bell_17
```

```python
df_bell_17.loc[38,:]=["서울도봉경찰서",622]
df_bell_17.loc[39,:]=["서울서초경찰서",0]
df_bell_17.loc[40,:]=["서울송파경찰서",815]
```

```python
df_bell_17 = df_bell_17.sort_values("관할서").reset_index(drop=True)
df_bell_17["date"]="2017"
df_bell_17
```

## 2018년도 비상벨 개수

```python
#18년도 비상벨 개수 : 총누적(20년도현황)- 20,19년도에 설치한 비상벨개수
#서초경찰서  추가됨
df_18 = df[-df["bell_set_yr"].isin([2020,2019])]
df_18
```

```python
#관할서별 비상벨 개수 데이터프레임화
df_bell_18 = pd.DataFrame(df_18.groupby("관할서", sort = True).count()["bell_us"]).reset_index()
df_bell_18.rename(columns={"bell_us":"bell_cnt"}, inplace=True)
df_bell_18
```

```python
# 송파경찰서는 17년도에 815개, 18년도에 834, 19년도에 864, 20년도에 879 ->강남구 추이랑 비슷하게함
# 도봉경찰서는 모든년도에 622 -> 강북, 노원구는 추이에 변화가 없다

df_bell_18.loc[39,:]=["서울도봉경찰서",622]
df_bell_18.loc[40,:]=["서울송파경찰서",834]
```

```python
df_bell_18 = df_bell_18.sort_values("관할서").reset_index(drop=True)
df_bell_18["date"]="2018"
df_bell_18
```

## 19년도 비상벨개수

```python
#19년도 비상벨 개수 : 총누적(20년도현황)- 20년도에 설치한 비상벨개수
df_19 = df[df["bell_set_yr"]!=2020]
df_19
```

```python
#관할서별 비상벨 개수 데이터프레임화
df_bell_19 = pd.DataFrame(df_19.groupby("관할서", sort = True).count()["bell_us"]).reset_index()
df_bell_19.rename(columns={"bell_us":"bell_cnt"}, inplace=True)
df_bell_19
```

```python
# 송파경찰서는 17년도에 815개, 18년도에 834, 19년도에 864, 20년도에 879 ->강남구 추이랑 비슷하게함
# 도봉경찰서는 모든년도에 622 -> 강북, 노원구는 추이에 변화가 없다

df_bell_19.loc[39,:]=["서울도봉경찰서",622]
df_bell_19.loc[40,:]=["서울송파경찰서",64]
```

```python
df_bell_19 = df_bell_19.sort_values("관할서").reset_index(drop=True)
df_bell_19["date"]="2019"
df_bell_19
```
