```python
import pandas as pd
```

# 데이터 전처리

```python
#파일불러오기
df = pd.read_csv("../raw_data/5.112신고.csv")
df.head()
```

```python
#데이터 확인
df.info()
```

```python
#결측치 파악
df.isnull().sum()
```

```python
#날짜데터 float -> str로 조정
df["date"] = df["date"].astype(int).astype(str)
```

```python
#컬럼명 변경
df.rename(columns = {"jur_stn":"관할서"},inplace = True)
```

```python
# 관할서에 "경찰서" 붙이기
df["관할서"] = df["관할서"]+"경찰서"
df.head()
```

```python
# 관할서명  수정
df["관할서"] = df["관할서"].str.replace("마산", "경남마산")                                     
df["관할서"] = df["관할서"].str.replace("창원", "경남창원")
df["관할서"] = df["관할서"].str.replace("수원", "경기수원")
df["관할서"] = df["관할서"].str.replace("진주", "경남진주")
df["관할서"] = df["관할서"].str.replace("진해", "경남진해")
df["관할서"] = df["관할서"].str.replace("세종", "충남세종")                                           
```

```python
df.head()
```

```python
#연도별 데이터 생성 및 인덱스 재설정
df_2018 = df[df["date"].str.contains("2018")].reset_index(drop = True)
df_2019 = df[df["date"].str.contains("2019")].reset_index(drop = True)
df_2020 = df[df["date"].str.contains("2020")].reset_index(drop = True)
df_2021 = df[df["date"].str.contains("2021")].reset_index(drop = True)
```

```python
#2018, 관할서별 신고수 파악
df_2018_dec = pd.DataFrame(df_2018.groupby(by="관할서", sort=True).count()["date"]).reset_index()
df_2018_dec
```

```python
#2019, 관할서별 신고수 파악
df_2019_dec = pd.DataFrame(df_2019.groupby(by="관할서", sort=True).count()["date"]).reset_index()
df_2019_dec
```

## 20년도 신고수 데이터는 추후 test를 위해 따로 저장

```python
## 근데 20년도는 세종경찰서 데이터가  없다...
df_2020[df_2020["관할서"].str.contains("충남세종경찰서")]
```

```python
#2020, 관할서별 신고수 파악
df_2020_dec = pd.DataFrame(df_2020.groupby(by="관할서", sort=True).count()["date"]).reset_index()
df_2020_dec

df_2020_dec.to_csv("../20년도_예측을위한_변수data/20년도_관할서별_신고수")
```

## 18,19병합

```python
# 병합
df_dec_merge = pd.merge(df_2018_dec, df_2019_dec, on=["관할서"]).rename(columns = {"date_x":"18_신고수","date_y":"19_신고수"})
df_dec_merge
```

```python
#파일저장
df_dec_merge.to_csv("전처리완성본/5_관할서별_신고수")
```
