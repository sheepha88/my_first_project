```python
import pandas as pd
```

```python
df = pd.read_csv("../raw_data/6.범죄발생원표.csv")
df.tail()
```

```python
##17년도 이전 데이터가 있는 것 확인
##17,18,19 년도만 추출
df = df[df["crm_date"].astype(str).str.contains("2017|2018|2019")]
df.head()
```

```python
df.info()
```

## 1. 날짜데이터 통일

```python
#편하게 보기 위해 str변환
date = df["crm_date"].astype(str)
date
```

```python
##"/" 없는 값에 "/"추가 한후, 변경
z = []
for i in date[-date.str.contains("/")]:
    b = list(i)
    b.insert(4,"/")
    b.insert(7,"/")
    a = "".join(b)
    z.append(a)
    
date[-date.str.contains("/")] = z
```

```python
#date값 "/"로 통일
date = date[date.str.contains("/")]
date
```

```python
#df의 날짜데이터 변경 후 다시 str로 변경(분석시 편함)
df["crm_date"] = pd.to_datetime(date)

#index 재조정
df.reset_index(drop=True, inplace = True)
df.tail()
```

```python
#df["crm_date"][0].strftime("%Y%m%d")

num = len(df)
z = []
for i in range(num):
    a = df["crm_date"][i].strftime("%Y%m%d")
    z.append(a)
    
z
```

```python
df["crm_date"] = z
df.head()
```

```python
df.to_csv("6_범죄발생원표_날짜전처리_완성본")
```

## 2. 연도별 정리

```python
df = pd.read_csv("../전처리완성본/범죄발생원표,범죄검거원표_전처리완성본/6_범죄발생원표_날짜전처리_완성본", index_col = 0)
```

```python
df
```

```python
df["crm_date"] = df["crm_date"].astype(str)
```

```python
# 연도만 가져오기
df.loc[df[df["crm_date"].str.contains("2017")]["crm_date"].index,"crm_date"]="2017"
df.loc[df[df["crm_date"].str.contains("2018")]["crm_date"].index,"crm_date"]="2018"
df.loc[df[df["crm_date"].str.contains("2019")]["crm_date"].index,"crm_date"]="2019"
```

```python
df["crm_date"].unique()
```

```python
#관할서 공백으로 인한 전처리 필요성 확인
df["jur_stn"] = df["jur_stn"].str.strip()
```

---

## 3. 관할서별 절도 폭력 발생건수

```python
df.head()
```

```python
# 연도별 , 관할서별 절도,폭력  발생 건수(관할서  가나다순 정리)
gr_절폭 = pd.DataFrame(df[df["5m_crm_yn"].isin(["절도","폭력"])].groupby(["jur_stn","crm_date"]).count()["crm"]).reset_index()
df_절폭 = gr_절폭.sort_values(["jur_stn","crm_date"]).rename(columns = {"crm":"절도폭력_발생수"})
df_절폭
```

## 4. 관할서별 강도 살인 발생건수

```python
# 관할서별 강도,살인  발생 건수(관할서  가나다순 정리)
gr_강살 = pd.DataFrame(df[df["5m_crm_yn"].isin(["강도","살인"])].groupby(["jur_stn","crm_date"]).count()["crm"]).reset_index()
df_강살 = gr_강살.sort_values(["jur_stn","crm_date"]).rename(columns = {"crm":"강도살인_발생수"})
df_강살
```

## 5. 연도별, 관할서별 교통사고 발생건수

```python
# 연도별 , 관할서별 절도,폭력  발생 건수(관할서  가나다순 정리)
gr_교통 = pd.DataFrame(df[df["crm"].str.contains("교통|도로|사고|도주차량")].groupby(["jur_stn","crm_date"]).count()["crm"]).reset_index()
df_교통 = gr_교통.sort_values(["jur_stn","crm_date"]).rename(columns = {"crm":"교통사고_발생수"})
df_교통
```

## 6. 법질서준수도( 기초질서, 집회지위질서 등 -> 5대범죄, 교통사고 제외)


```python
gr_질서 = df[df["5m_crm_yn"]==' ']
gr_질서 = gr_질서[-gr_질서["crm"].str.contains("교통|도로|사고|도주차량")].groupby(["jur_stn","crm_date"]).count()["crm"].reset_index()
df_질서 = gr_질서.sort_values(["jur_stn","crm_date"]).rename(columns = {"crm":"법질서_발생수"})
df_질서
```

## 7. 전반적 안전도(ALL, 모든범죄건수)

```python
#법질서준수도
gr_all = pd.DataFrame(df.groupby(["jur_stn","crm_date"]).count()["crm"]).reset_index()
df_all = gr_all.sort_values(["jur_stn","crm_date"]).rename(columns = {"crm":"전반적_발생수"})
df_all
```

----
## 8. 모든파일 병합

```python
df_범죄발생원표 = pd.concat([df_절폭, df_강살.iloc[:,-1] , df_교통.iloc[:,-1], df_질서.iloc[:,-1] , df_all.iloc[:,-1]], axis = 1)
df_범죄발생원표.head()
```

## 9. 전처리 완성본 파일 저장

```python
df_범죄발생원표.to_csv("../전처리완성본/6_범죄발생원표_전처리완성본")
```
