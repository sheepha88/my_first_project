```python
import pandas as pd
```

```python
# 파일 불러오기
df_발생 = pd.read_csv("../전처리완성본/6_범죄발생원표_전처리완성본", index_col=0)
df_검거 = pd.read_csv("../전처리완성본/7_범죄검거원표_전처리완성본", index_col=0)
```

## 1. 생성할 데이터프레임 컬럼 준비

```python
#데이터프레임 컬럼변경 검거->검거율
col = df_검거.columns[0:].str.replace("검거수", "검거율")
col
```

## 2. 검거율 계산 및 데이터프레임  생성

```python
# nuumpy로 계산
#데이터  프레임  생성
import numpy as np
df = pd.DataFrame(np.array(df_검거.iloc[:,2:])/np.array(df_발생.iloc[:,2:]))


# 소수점  셋째자리 까지 표시
df = df.round(3)

df = pd.concat([df_검거.iloc[:,:2], df], axis = 1)
df.columns = col
```

```python
df.head(50)
```

##  3.데이터전처리 완성본 파일 저장

```python
df.to_csv("../전처리완성본/0_관할서별_연도별_검거율현황")
```

### +다른방법

```python
#반복문으로 검거율 계산
result = []
for i,z in zip(df_검거.columns[1:] , df_발생.columns[1:]):
    num = df_검거[i] / df_발생[z]
    result.append(num)
    
result[0]
```

```python
# 컬럼하고 검거율 딕트로 묶기 ->데이터프레임 생성을 위한 밑작업
df_dict = dict(zip(col, result))

# 관할서 딕트 추가
df_dict["관할서"] = df_검거["관할서"]
```

```python
#데이터 프레임 생성
df_검거율 = pd.DataFrame(df_dict)
```

```python
# 컬럼 순서 변경
ch_cols = ['관할서','검거율_17_절도폭력', '검거율_18_절도폭력', '검거율_19_절도폭력', '검거율_17_강도살인',
       '검거율_18_강도살인', '검거율_19_강도살인', '검거율_17_교통사고', '검거율_18_교통사고',
       '검거율_19_교통사고', '검거율_17_기초질서', '검거율_18_기초질서', '검거율_19_기초질서', '검거율_17_전반',
       '검거율_18_전반', '검거율_19_전반']

df_검거율 = df_검거율[ch_cols]
```

```python
# 소수점 셋째자리 까지 반올림
# 완성본
df_검거율 = df_검거율.round(3)
df_검거율
```
